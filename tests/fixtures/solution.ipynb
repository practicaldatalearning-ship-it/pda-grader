# House prices — solution (synthetic)

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.read_csv('train.csv')
```

```python
# BEGIN SOLUTION
n_missing = int(df.isna().sum().sum())
# END SOLUTION
```

```python
# BEGIN SOLUTION
corrs = df.drop(columns=['id']).corr(numeric_only=True)['price'].drop('price').abs()
top_feature = corrs.idxmax()
# END SOLUTION
```

```python
# BEGIN SOLUTION
train = df.dropna()
X, y = train[['area', 'rooms']], train['price']
model = LinearRegression().fit(X, y)
full = df.copy()
full['area'] = full['area'].fillna(full['area'].mean())
full['rooms'] = full['rooms'].fillna(full['rooms'].median())
preds = pd.DataFrame({'id': full['id'], 'price': model.predict(full[['area', 'rooms']])})
preds.to_csv('predictions.csv', index=False)
# END SOLUTION
```

```python
# BEGIN SOLUTION
explanation = 'Area is the strongest predictor of price: larger homes command higher prices, and it has the highest correlation with the target.'
# END SOLUTION
```
